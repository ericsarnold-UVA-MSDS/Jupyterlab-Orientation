Given a random set of data, Normal Distributions and Binomial Distributions rely on a common set of values for a large N value, which represents the number of trials or occurances.
We will show the data for a small number of trials, then show that as N increases, Binomial distributions begin to converge.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def NormDist(x,amp,mu,sigma):
    return amp*np.exp(-(x-mu)**2/(2*sigma**2))
N= 10
r= np.empty([N])
for i in range(10):
    r += np.random.randint(1,7,N)
x= np.linspace(10,60,50)
n,bins,patches = plt.hist(r,bins=50,range=(10,60))
y = n
plt.title('Normal Distribution of Ten Dice rolled 10 times')
plt.xlabel('Number Rolled')
plt.ylabel('Entries/Frequency')
plt.show(NormDist)
```

Five, six sided fair dice are rolled simultaneously and the values of each is summed to give a value between 10 and 60. For a N of 10 trials, there is no apparent bell-curve shape or trend in the data set.

```python
n = 10
p = .5 #Probability of flipping fair coin is 50/50
N = 10
BinomDist = np.random.binomial(n,p,N)
plt.hist(BinomDist,bins=10,color='red')
plt.title('Binomial Distribution of flipping 10 coins 10 times')
plt.xlabel('Coin Flips')
plt.ylabel('Entries/Frequency')
plt.show()
```

Five fair coins are flipped and the flips and the number of heads and tails is demonstrated as a binomial distribution. With only ten trials (N) of ten coin flips recorded, you can see there is no trend.

If we increase the N to approach infinity, we will see the two data sets converge. See the transition as we complete 10x, 100x and 1000x trials per example above.
