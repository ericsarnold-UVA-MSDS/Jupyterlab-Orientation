Dice Rolls (Nromal):

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def NormDist(x,amp,mu,sigma):
    return amp*np.exp(-(x-mu)**2/(2*sigma**2))

N= 100000
r= np.empty([N])

for i in range(5):
    r += np.random.randint(1,7,N)

x= np.linspace(5,30,25)

n,bins,patches = plt.hist(r,bins=25,range=(5,30))

y = n

plt.title('Normal Distribution of Five Dice rolled')
plt.xlabel('Number Rolled')
plt.ylabel('Entries/Frequency')
plt.show(NormDist)
```

Coin Flips (Binomial):

```python
n = 10
p = .5 #Probability of flipping fair coin is 50/50
N = 10000

BinomDist = np.random.binomial(n,p,N)

plt.hist(BinomDist,bins=10,)

plt.title('Binomial Distribution of flipping 10 coins 10000 times')
plt.xlabel('Coin Flips')
plt.ylabel('Entries/Frequency')
plt.show()
```

Crashes per day (Poisson):

```python
lam = 3.5
size = 10000

PoissonDist = np.random.poisson(lam, size)

plt.hist(PoissonDist)

plt.title('Poissons Distribution for crashes at a busy intersection per day')
plt.xlabel('Crashes per day')
plt.ylabel('Frequency')
plt.show()
```
